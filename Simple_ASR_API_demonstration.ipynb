```python
# Install necessary libraries
!pip install SpeechRecognition

# Import libraries
import speech_recognition as sr
import io
from IPython.display import Audio, display
from google.colab import files
```

```python
# Upload an audio file
uploaded = files.upload()

# Load the audio file
file_name = next(iter(uploaded))  # Get the uploaded file name
display(Audio(file_name, autoplay=True))

# Initialize speech recognizer
recognizer = sr.Recognizer()

# Convert AudioFile to AudioData
with sr.AudioFile(file_name) as source:
    audio_data = recognizer.record(source)

# Recognize the speech using Google's Web Speech API
try:
    text = recognizer.recognize_google(audio_data)
    print("Transcription: " + text)
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand audio")
except sr.RequestError as e:
    print(f"Could not request results from Google Speech Recognition service; {e}")
```

```python
!pip install -U openai-whisper
# Install the OpenAI module
!pip install openai==0.28
!pip install openai pydub

from pydub import AudioSegment
import openai
from IPython.display import Audio, display

# Upload an audio file
uploaded = files.upload()

# Load the audio file
file_name = next(iter(uploaded))  # Get the uploaded file name
display(Audio(file_name, autoplay=True))

# Convert the WAV file to MP3
mp3_file_name = file_name.replace('.wav', '.mp3')
audio = AudioSegment.from_wav(file_name)
audio.export(mp3_file_name, format='mp3')
print(f"Converted {file_name} to {mp3_file_name}")

# Load your OpenAI API key
openai.api_key = 'sk--xxxxx'  # transfer your OpenAI API key

# Use OpenAI's Whisper ASR model to transcribe the MP3 audio
try:
    with open(mp3_file_name, 'rb') as audio_file:
        response = openai.Audio.transcribe(model="whisper-1", file=audio_file)
    text = response['text']
    print("Transcription: " + text)
except Exception as e:
    print(f"Could not transcribe audio using Whisper model; {e}")
```
